```python
df_transfers_out = df_transfers[['hash', 'from', 'value', 'metadata.blockTimestamp']].copy()
df_transfers_out['value'] = df_transfers_out['value'].apply(lambda x: -1.0* x)
df_transfers_out.rename(columns = {'hash': 'tx_hash', 'from':'address', 'value':'amount', 'metadata.blockTimestamp':'tx_timestamp'}, inplace=True)
df_transfers_out.head()
```

```python
df_transfers_in = df_transfers[['hash', 'to', 'value', 'metadata.blockTimestamp']].copy()
df_transfers_in.rename(columns = {'hash': 'tx_hash', 'to':'address', 'value':'amount', 'metadata.blockTimestamp':'tx_timestamp'}, inplace=True)
df_transfers_in.head()
```

```python
# transfer_in_n_out = pd.concat([df_transfers_in.head(), df_transfers_out.head()], axis=0)
transfer_in_n_out=pd.concat([df_transfers_in, df_transfers_out], axis=0)
transfer_in_n_out['date'] = pd.to_datetime(transfer_in_n_out['tx_timestamp'].apply(lambda x:x.date()))
transfer_in_n_out
```

```python
transfers_all = transfer_in_n_out[['date', 'address', 'amount']]
```

```python
import math
len_wallet_mid = round(len(wallets)/2)
len_wallet = len(wallets)

wallets1 = wallets[0:len_wallet_mid-1][0].to_list()
wallets2 = wallets[len_wallet_mid: len_wallet-1][0].to_list()

wallet_list = [wallets1, wallets2]
```

```python
wallet_bals = wallet_all_days_n_changes.sort_values('date', ascending=True)

wallet_bals1 = wallet_bals[wallet_bals['wallet_address'].isin(wallets1)]
wallet_bals2 = wallet_bals[wallet_bals['wallet_address'].isin(wallets2)]

wallet_bals1_sum = wallet_bals1.groupby(by=['date', 'wallet_address'])['amount'].sum().reset_index()
wallet_bals2_sum = wallet_bals2.groupby(by=['date', 'wallet_address'])['amount'].sum().reset_index()

# wallet_bals = pd.concat([wallet_bals1, wallet_bals2], axis=0)
    

# wallet_bals['net_change'] = wallet_bals.groupby(by=['date','wallet_address'])['amount'].sum().reset_index()
# wallet_all_days_n_changes['cumsum'] = wallet_all_days_n_changes.groupby(['date','wallet_address'])['net_change'].cumsum()
# wallet_bals
```

```python
wallet_bals1_sum['cumsum'] = wallet_bals1_sum.groupby(['wallet_address'])['amount'].cumsum()
wallet_bals2_sum['cumsum'] = wallet_bals2_sum.groupby(['wallet_address'])['amount'].cumsum()
wallet_bals = pd.concat([wallet_bals1_sum, wallet_bals2_sum], axis=0)
wallet_bals = wallet_bals[wallet_bals['cumsum']>=5].sort_values(by=['date', 'cumsum'], ascending=[False, False])
wallet_bals.to_csv(f'{date.today()}_balances_by_day.csv')
```
